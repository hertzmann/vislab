```python
import numpy as n
def read_numpy_data_batch(batch_path):
        file = open(batch_path, 'rb')
        np_batch = n.load(file)
        batch = {}
        try:
            batch['batch_label'] = np_batch['batch_label'][0]
        except:
            batch['batch_label'] = ''
        batch['data'] = n.require(np_batch['data'], dtype=n.single, requirements='C')
        batch['labels'] = np_batch['labels']
        try:
            batch['filenames'] = np_batch['filenames'].tolist()
        except:
            batch['filenames'] = [''] * batch['labels'].shape[0]
        file.close()
        return batch
```
