```python
%load_ext autoreload
%autoreload 2 

import re
import requests
import bs4
import aphrodite.dataset
import cPickle
import os
import networkx as nx
import time
from networkx.algorithms import bipartite
```

```python
ava_df = aphrodite.dataset.load_ava_df()
```

```python
client = aphrodite.util.get_mongodb_client()
collection = client.ava_feat_db.fav_user_ids
collection.count()
```

```python
pickle_filename = '../data/fav_user_ids_graph.pickle'
force = True

if force or not os.path.exists(pickle_filename):
    B = nx.Graph()
    t = time.time()
    for doc in collection.find().limit(10000):
        image_id = 'i_{}'.format(doc['image_id'])

        connections = cPickle.loads(doc['feat'])
        if len(connections) == 0:
            continue
            
        B.add_node(image_id, bipartite=0)
        for user_id, date in connections:
            user_id = 'u_{}'.format(user_id)
            B.add_node(user_id, bipartite=1)
            B.add_edge(image_id, user_id, date=date)
    
    print("Finished in {:.3f} s".format(time.time() - t))
    image_nodes = set(n for n,d in B.nodes(data=True) if d['bipartite']==0)
    user_nodes = set(B) - image_nodes
    
    graph_data = {'B': B, 'image_nodes': image_nodes, 'user_nodes': user_nodes}
    
    with open(pickle_filename, 'w') as f:
        cPickle.dump(graph_data, f, protocol=2)

else:
    graph_data = cPickle.load(open(pickle_filename, 'rb'))

B, image_nodes, user_nodes = graph_data['B'], graph_data['image_nodes'], graph_data['user_nodes']
```

```python
print('Nodes (total/users/images): {}/{}/{}'.format(
    len(B.nodes()), len(image_nodes), len(user_nodes)))
print('Edges: {}'.format(len(B.edges())))
```
