```python
import caffe
import os
import skimage.io
import skimage.transform
import numpy as np
import pandas as pd
import vislab.datasets
%matplotlib inline

#from matplotlib import pyplot
import scipy.interpolate

import cma   # install with "pip install cma"
#import PIL
#from PIL import ImageEnhance
```

```python
#sourceImageFilename = '/Users/hertzman/Desktop/sk/images/cal.jpg'
caffeDir = '/Users/hertzman/Desktop/sk/caffe'
modelDir = '/Users/hertzman/Desktop/sk/data'

sourceImageFilename = caffeDir + '/examples/images/cat.jpg'

def load_image(filename):
    img = skimage.io.imread(filename)
    if img.ndim == 2:
        img = np.tile(img[:, :, np.newaxis], (1, 1, 3))
    elif img.shape[2] == 4:
        img = img[:, :, :3]
    return img


sourceImage = load_image(sourceImageFilename)

skimage.io.imshow(sourceImage)
```

```python
IMAGE_DIM = 227
IMAGE_MEAN = np.load(caffeDir+'/python/caffe/imagenet/ilsvrc_2012_mean.npy')
PIXEL_MEAN = IMAGE_MEAN.mean(0).mean(0)[np.newaxis, np.newaxis, :]

def get_image_blob(image, window=None):
    """
    Input
    -----
    image: (H x W x 3) ndarray
    window: (4) ndarray
        (ymin, xmin, ymax, xmax) coordinates, 0-indexed

    Output
    ------
    image: (3 x H x W) ndarray
        Resized to IMAGE_DIM.
    """
    # Crop a subimage if window is provided.
    if window is not None:
        image = image[window[0]:window[2], window[1]:window[3]]

    # Resize to input size, subtract mean, convert to BGR.
    image = image[:, :, ::-1]
    image = skimage.transform.resize(image, (IMAGE_DIM, IMAGE_DIM)) * 255
    image -= PIXEL_MEAN
    image = image.swapaxes(1, 2).swapaxes(0, 1)
    
    # Format for immediate use in caffe.
    blob = np.ascontiguousarray(image[np.newaxis, :, :, :]).astype(np.float32)
    return blob

model_def_file = modelDir + '/flickr_finetune_deploy.prototxt'
pretrained_model = modelDir + '/flickr_finetune_train_iter_3000'
caffenet = caffe.Net(model_def_file, pretrained_model)
caffenet.set_phase_test() 

style_names = vislab.datasets.flickr.underscored_style_names

NUM_CLASSES = 20
output_blobs = [np.empty((1, NUM_CLASSES, 1, 1), dtype=np.float32)]

def get_preds(image):
    caffenet.Forward([get_image_blob(image)], output_blobs)
    feats = [output_blobs[0][i].flatten() for i in range(len(output_blobs[0]))]
    preds = pd.Series(feats[0], style_names)
    preds.sort(ascending=False)
    return preds
```
