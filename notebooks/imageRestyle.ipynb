```python
def toyTransform(params):

    # from looking at the source code, the first three are pixelwise transforms that are linear in RGB, 
    # and the last is a simple image sharpening (it's not very stable, leads to ringing very quickly)

    im0 = PIL.ImageEnhance.Color(sourceImage).enhance(params[0])
    im1 = PIL.ImageEnhance.Contrast(im0).enhance(params[1])
    im2 = PIL.ImageEnhance.Brightness(im1).enhance(params[2])
    im3 = PIL.ImageEnhance.Sharpness(im2).enhance(params[3])

    return im3


pyplot.imshow(toyTransform([0,1,1.5,1]))
toyTransformIdentityParams = [1,1,1,1]
```

```python
def varianceObjective(params):
#    targetVar = np.array([4357.002770470229, 4358.939853533155, 3293.2595291003545])
    targetVar = np.array([4000,4500,3200])
    
    newVar = np.array(PIL.ImageStat.Stat(toyTransform(params)).var)

    return np.linalg.norm(targetVar - newVar)

print(varianceObjective([1,1,1,1]))
```

```python
initParams = [1,1,1,1]  # identity transform
result = cma.fmin(varianceObjective, x0=initParams, sigma0=1, maxfevals = 1000)
```

```python
optParams = result[0]
print('Opt solution: '+str(optParams))
resultImage = toyTransform(optParams)
print(PIL.ImageStat.Stat(resultImage).var)
pyplot.imshow(resultImage)
```
